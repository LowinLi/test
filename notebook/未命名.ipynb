```python
import os
import sys
import gensim
import pandas as pd
```

```python

# 加载词向量文件
#model = gensim.models.word2vec.Word2Vec.load("sgns.weibo.word")
model = gensim.models.KeyedVectors.load_word2vec_format("sgns.weibo.word")
```

```python
labels = '快乐 乐观 平静 担忧 无助 恐惧 悲伤 愤怒 恐慌 压力 抑郁 焦虑'.split(' ')
for label in labels:
    tt = []
    for i in model.most_similar(label,topn=5000):
        tt.append((i[0],i[1]))
    df = pd.DataFrame(tt)
    df.to_excel(label+'.xlsx')
```
