```python
import pymongo
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import copy
client = pymongo.MongoClient('localhost',27017)
start = datetime(2019,12,13)
end = datetime(2020,3,20)
```

```python
co = client.Weibo_Local_Db.comment
datas = {}
for label in '全部 快乐 乐观 平静 无助 恐惧 悲伤 愤怒 恐慌 压力 抑郁 焦虑'.split(' '):
    datas[label]={}
    print(label)
    start_one = copy.deepcopy(start)
    while start_one<end:
        end_one = start_one+timedelta(days=1)
        print(start_one, end_one)
        query = {'comment_dt':
                 {'$gt':start_one,
                  '$lte':end_one
                 }
                }
        if label != '全部':
            query[label]={'$exists':True}
        count = co.find(query).count()
        print(count)
        datas[label][start_one.strftime('%Y%m%d')]=count
        start_one += timedelta(days=1)
```
